```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline
```

```python
# step size
h = 0.1
# define grid
x = np.arange(0, 2*np.pi, h)
# compute function
y = np.cos(x)

# compute vector of foward differences
foward_diff = np.diff(y)/h
# compute corresponding gird
x_diff = x [:-1]
# compute exact solution
exact_solution = -np.sin(x_diff)

# plot solution
plt.figure(figsize = (12, 8))
plt.plot(x_diff, foward_diff, '--', \
         label = 'Finate difference approximation')
plt.plot(x_diff, exact_solution, \
         label = 'Exact solution')
plt.legend()
plt.show()

# compute max eror between
# numerical devirative and exact solution
max_eror = max(abs(exact_solution - foward_diff))
print(max_eror)
```
